# Gráfico de barra

```python
from matplotlib import pyplot as plt
from collections import Counter
```

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablaca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
```

```python
'''
Barras possuem o tamanho padrão de 0.8, então adicionaremos 0.1 ás
coordenadas à esquerda para que cada barra seja centralizada
'''
xs = [i + 0.1 for i, _ in enumerate(movies)]
```

```python
# as barras do gráfico com as coordenas x á esquerda [xs], alturas [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de Premiações")
plt.title("Meus Filmes Favoritos")

# nomeia o eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```

### Gráfico histograma de valores numéricos

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()], # move cada para a esquerda em 4
       histogram.values(), # dá para cada barra sua altura correta
       8) # dá para cada barra a largura de 8

plt.axis([-5, 105, 0, 5]) # eixo x de -5 até 105, eixo y de 0 até 5

plt.xticks([10 * i for i in range(11)]) # rótulos de eixo x em 0, 10, ..., 100
plt.xlabel('Decil')
plt.ylabel('# de Alunos')
plt.title('Distribuição das Notas do Teste 1')

plt.show()
```
