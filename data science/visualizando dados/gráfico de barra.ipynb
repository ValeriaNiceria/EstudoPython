# Gráfico de barra

```python
from matplotlib import pyplot as plt
```

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablaca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
```

```python
'''
Barras possuem o tamanho padrão de 0.8, então adicionaremos 0.1 ás
coordenadas à esquerda para que cada barra seja centralizada
'''
xs = [i + 0.1 for i, _ in enumerate(movies)]
```

```python
# as barras do gráfico com as coordenas x á esquerda [xs], alturas [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de Premiações")
plt.title("Meus Filmes Favoritos")

# nomeia o eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```
