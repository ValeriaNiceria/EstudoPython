# Ferramentas Funcionais

```python
def exp(base, power):
    return base ** power
```

```python
def two_to_the(power):
    return exp(2, power)

two_to_the(4)
```

```python
from functools import partial
two_to_the = partial(exp, 2) # agora é uma função de uma variável
print(two_to_the(3))
```

```python
square_of = partial(exp, power=2)
print(square_of(3))
```

```python
def double(x):
    return 2 * x

xs = [1, 2, 3, 4]
twice_xs = [double(x) for x in xs]
print(twice_xs)
```

## map

```python
twice_xs = map(double, xs)
list(twice_xs)
```

```python
list_doubler = partial(map, double)
twice_xs = list_doubler(xs)
list(twice_xs)
```

```python
def multiply(x, y): return x * y

products = map(multiply, [1, 2], [4, 5])

list(products)
```

## filter

```python
def is_even(x):
    # True se x for par, False se x for ímpar
    return x % 2 == 0

x_evens = [x for x in xs if is_even(x)]
print(x_evens)
```

```python
x_evens = filter(is_even, xs)

list(x_evens)
```

```python
list_evener = partial(filter, is_even)
x_evens = list_evener(xs)

list(x_evens)
```

## reduce

```python
# Importando a função reduce do módulo functools
from functools import reduce


x_product = reduce(multiply, xs)
x_product
```

```python
list_product = partial(reduce, multiply) # função que reduz uma lista
x_product = list_product(xs)
x_product
```
