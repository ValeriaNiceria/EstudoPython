# Funções

```python
def double(x):
    return x * 2
```

```python
double(3)
```

```python
def apply_to_one(f):
    # chama a função com 1 como seu argumento
    return f(1)

my_double = double # refere-se à função definida anteriormente
x = apply_to_one(my_double) # é igual a 2
x
```

```python
y = apply_to_one(lambda x: x + 4) # é igual a 5
y
```

```python
another_double = lambda x: 2 * x
another_double(2)
```

```python
def another_double(x): return 2 * x
another_double(4)
```

```python
def my_print(message="my default message"):
    print(message)
```

```python
my_print('hello')
```

```python
my_print()
```

```python
def subtract(a=0, b=0):
    return a - b
```

```python
subtract(10, 5)
```

```python
subtract(0, 5)
```

```python
subtract(b=5)
```
