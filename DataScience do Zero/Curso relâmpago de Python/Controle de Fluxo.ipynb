# Controle de Fluxo

```python
if 1 > 2:
    print('if only 1 werw greater than two...')
elif 1 > 3:
    print('elif stands for "else if"')
else:
    print('when all else fails use else (if you want to)')
```

```python
# ternário if-then-else
parity = "even" if 10 % 2 == 0 else "odd"

parity
```

```python
# loop while
x = 0
while x < 10:
    print(x, "is less than 10")
    x += 1
```

```python
for x in range(10):
    print(x, 'is less than 10')
```

```python
for x in range(10):
    if x == 3:
        continue # vai para a próxima iteração imediatamente
    if x == 5:
        break # sai do loop completamente
    print(x)
```
