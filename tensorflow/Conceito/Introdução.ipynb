# Introdução ao TensorFlow

O TensorFlow é uma das bibliotecas mais amplamente utilizadas para implementar o aprendizado de máquina e outros algoritmos que envollvem grandes operações matemáticas. O TensorFlow foi desenvolvido pelo Google e é uma das bibliotecas de aprendizado de máquina mais populares no GitHub. O Google usa o TensorFlow para aprendizado de máquina em quase todos os aplicativos. Como por exemplo o Google Photos ou o Google Voice Search.

Matematicamente, um tensor é um vetor N-dimensional, significando que um tensor pode ser usado para representar conjuntos de dados N-dimensionais.

Os tensores podem ser constantes ou variáveis.
