```python
# Dicionário de listas
dic3 = {'key1': 1234, 'key2': [20, 30, 49], 'key3': ['leite', 'carne',  'batata']}
```

```python
dic3
```

```python
dic3['key2']
```

```python
# Acessando um item da lista, dentro do dicionário
dic3['key3'][0].upper()
```

```python
# Operações com itens da lista, dentro do dicionário
var1 = dic3['key2'][0] - 2
```

```python
var1
```

```python
# Duas operações no mesmo comando, para atualizar um item dentro da lista
dic3['key2'][0] -= 2
```

```python
dic3
```

## Criando dicionários aninhados

```python
# Criando dicionários aninhados
dicAninhado = {'key1': {'key2_aninhada': {'key3_aninhada': 'Dicionário aninhado em Python'}}}
```

```python
dicAninhado
```

```python
dicAninhado['key1']['key2_aninhada']['key3_aninhada']
```
