# Função Map

```python
# Criando duas funções

'''
Função 1:
Recebe uma temperatura como parâmetro e retorna a temperatura em Fahrenheit
'''
def fahrenheit(T):
    return ((float(9) / 5) * T + 32)

'''
Função 2:
Recebe uma temperatura como parâmetro e retorna a temperatura em Celsius
'''
def celsius(T):
    return (float(5) / 9) * (T - 32)
```

```python
# Criando uma lista
temperaturas = [0, 22.5, 40, 100]
```

```python
# Aplicando a função a cada elemento da lista de temperaturas
# Em Python 3, a função map() retorna um interator
map(fahrenheit, temperaturas)
```

```python
# Função map() retorna uma lista de temperaturas convertidas em Fahrenheit
list(map(fahrenheit, temperaturas))
```

```python
# Usando um loop for para imprimir o resultado da função map()
for temp in map(fahrenheit, temperaturas):
    print(temp)
```

```python
# Convertendo para Celsius
list(map(celsius, temperaturas))
```

```python
# Usando o loop para percorrer cada valor
for temp in map(celsius, temperaturas):
    print(temp)
```

```python
# Usando lambda
list(map(lambda x: (float(5) / 9) * (x - 32), temperaturas))
```

```python
# Somando os elementos de 2 listas
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

list(map(lambda x, y: x + y, a, b))
```

```python
# Somando os elementos de 3 listas
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]
c = [9, 10, 11, 12]

list(map(lambda x, y, z: x + y + z, a, b, c))
```
