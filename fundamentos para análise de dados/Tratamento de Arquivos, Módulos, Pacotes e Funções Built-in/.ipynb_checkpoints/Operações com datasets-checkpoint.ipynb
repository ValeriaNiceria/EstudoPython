## Automatizando o processo de gravação

```python
fileName = input('Digite o nome do arquivo: ')
```

```python
fileName = fileName + '.txt'
```

```python
arq3 = open(fileName, 'w')
```

```python
arq3.write('Incluíndo texto no arquivo criado')
```

```python
arq3.close()
```

```python
arq3 = open(fileName, 'r')
```

```python
print(arq3.read())
```

```python
arq3.close()
```
