# Função filter

```python
# Criando uma função
def verificarPar(num):
    if num % 2 == 0:
        return True
    else:
        return False
```

```python
'''
Chamando a funçãoe passando um número como parâmetro.
Retornará, falso se for ímpar e true se for par.
'''
verificarPar(45)
```

```python
lista = []
for x in range(19):
    lista.append(x)
```

```python
lista
```

```python
filter(verificarPar, lista)
```

```python
list(filter(verificarPar, lista))
```

```python
list(filter(lambda x: x % 2 == 0, lista))
```

```python
list(filter(lambda x: x > 8, lista))
```
