## Criando Matriz

```python
# Importando o módulo NumPy
import numpy as np
```

```python
# Criando uma matriz
matriz = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
print(matriz)
```

```python
matriz.shape
```

```python
# Criando uma matriz 2x3 apenas com números "1"
matriz1 = np.ones((2, 3))
```

```python
print(matriz1)
```

```python
# Criando uma matriz a partir de uma lista de listas
lista = [[13, 14, 15], [16, 17, 18], [19, 20, 21]]
```

```python
# A função matriz cria uma matriz a partir de uma sequência
matriz2 = np.matrix(lista)
```

```python
print(matriz2)
```

```python
type(matriz2)
```

```python
# Formato da matriz
np.shape(matriz2)
```

```python
matriz2.size
```

```python
matriz2.dtype
```

```python
matriz2.itemsize # tamanho que será ocupado na memória do computador
```

```python
matriz2.nbytes
```

```python
matriz2[2, 1]
```

```python
# Alterando um elemento da matriz
matriz2[1, 0] = 100
```

```python
matriz2
```

```python
x = np.array([1, 2]) # NumPy decide o tipo dos dados
y = np.array([1.0, 2.0]) # NumPy decide o tipo dos dados
z = np.array([1, 2], dtype = np.float64) # Forçamos um tipo de dados em particular

print(x.dtype, y.dtype, z.dtype)
```
