## Outras Operações com Arrays

```python
# Importando o NumPy
import numpy as np
```

```python
# Slicing
a = np.diag(np.arange(3))
```

```python
a
```

```python
a[1, 1]
```

```python
a[1]
```

```python
b = np.arange(10)
```

```python
b
```

```python
# [start:end:step]
b[2:9:3]
```

```python
# Comparação
a = np.array([1, 2, 3, 4])
b = np.array([4, 5, 6, 7])
a == b
```

```python
np.array_equal(a, b)
```

```python
a.min()
```

```python
a.max()
```

```python
# Somando um elemento ao array
np.array([1, 2, 3]) + 1.5
```

```python
# Usando o método around
a = np.array([1.2, 1.5, 1.6, 2.5, 3.5, 4.5])
```

```python
b = np.around(a)
```

```python
b
```

```python
# Criando um array
b = np.array([1, 2, 3, 4])
```

```python
b
```

```python
# Copiando um array
c = b.flatten()
```

```python
c
```

```python
# Criando um array
v = np.array([1, 2, 3])
```

```python
# Adicionando uma dimensão ao array
v[:, np.newaxis], v[:, np.newaxis].shape, v[np.newaxis, :].shape
```

```python
# Repetindo os elementos de um array
np.repeat(v, 3)
```

```python
# Repetindo os elementos de um array
np.tile(v, 3)
```

```python
# Criando um array
w = np.array([5, 6])
```

```python
w
```

```python
# Concatenando 
np.concatenate((v, w), axis=0)
```

```python
# Copiando arrays
r = np.copy(v)
```
