# Series

Series é um array unidimensional que contém um array de dados e um array de labels, chamado índice.

```python
from pandas import Series
```

```python
import pandas as pd
```

```python
pd.__version__
```

```python
# Criando uma séries sem especificar os índices
obj = Series([67, 78, -56, 13])
```

```python
obj
```

```python
type(obj)
```

```python
obj.values
```

```python
obj.index
```

```python
# Criando uma série e especificando os índices
obj2 = Series([67, 78, -56, 13], index = ['a', 'b', 'c', 'd'])
```

```python
obj2
```

```python
obj2.values
```

```python
obj2.index
```

```python
obj2[obj2 > 3]
```

```python
obj2['b']
```

```python
'd' in obj2
```

```python
# Criando uma série de dados passando um dicionário como parâmetro
dict = {'Futebol': 5200, 'Tênis': 120, 'Natação': 698, 'Volleyball': 1550}
```

```python
# Criando um série a partir de um dicionário
obj3 = Series(dict)
```

```python
obj3
```

```python
type(obj3)
```

```python
# Criando uma lista
esportes = ['Futebol', 'Tênis', 'Natação', 'Basktetball']
```

```python
# Criando uma serie e usando uma lista como índice
obj4 = Series(dict, index=esportes)
```

```python
obj4
```

```python
pd.isnull(obj4)
```

```python
pd.notnull(obj4)
```

```python
obj4.isnull()
```

```python
# Concatenando Series
obj3 + obj4
```

```python
obj4.name = 'população'
```

```python
obj4.index.name = 'esporte'
```

```python
obj4
```
