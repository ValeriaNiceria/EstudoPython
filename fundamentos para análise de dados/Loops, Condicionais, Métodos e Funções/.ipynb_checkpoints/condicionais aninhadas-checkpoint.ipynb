# Condicionais Aninhadas

```python
idade = 18
if idade >= 18:
    print('Você pode dirigir!')
```

```python
nome = 'Bob'
if idade > 13:
    if nome == 'Bob':
        print('Ok Bob, você foi autorizado a entrar!')
else:
    print('Desculpe, mas você não pode entrar!')
```

```python
idade = 13
if idade >= 13 and nome == 'Bob':
    print('Ok Bob, você foi autorizado a entrar!')
```

```python
idade = 12
if (idade >= 13) or (nome == 'Bob'):
    print('Ok Bob, você foi autorizado a entrar!')
```

# Elif

```python
dia = 'Terça'
if dia == 'Segunda':
    print('Hoje fará sol!')
else:
    print('Hoje vair chover!')
```

```python
if dia == 'Segunda':
    print('Hoje fará sol!')
elif dia == 'Terça':
    print('Hoje vai chover!')
else:
    print('Sem previsão do tempo para o dia selecionado.')
```
