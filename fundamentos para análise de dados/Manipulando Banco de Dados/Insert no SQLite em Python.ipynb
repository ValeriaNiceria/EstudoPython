# Criando o Banco de Dados e Inserindo Dados

```python
# Remove o aquivo com o banco de dados SQLite (caso exista)
import os
os.remove('dsa.db') if os.path.exists('dsa.db') else None
```

```python
import sqlite3

# Criando um conexão
con = sqlite3.connect('dsa.db')

# Criando um cursor
c = con.cursor()

# Função para criar uma tabela
def create_table():
    c.execute('CREATE TABLE IF NOT EXISTS produtos '\
             '(id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, '\
             'date TEXT, '\
             'prod_name TEXT, '\
             'valor REAL)')
    
# Função para inserir uma linha
def data_insert():
    c.execute('INSERT INTO produtos VALUES(10, "2019-03-19 21:35:19", "Teclado", 90)')
    con.commit()
    c.close()
    con.close()
```

```python
# Criar tabela
create_table()
```

```python
# Inserir dados
data_insert()
```
