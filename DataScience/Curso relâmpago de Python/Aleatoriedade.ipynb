# Aleatoriedade

```python
import random
```

```python
four_uniform_randoms = [random.random() for _ in range(2)]

four_uniform_randoms # random.random() produz números uniformemente entre 0 e 1
```

```python
# configurando com random.seed para obter resultados reproduzíveis
random.seed(10) # configura seed para 10
print(random.random())
```

```python
random.seed(10) # reinicia seed para 10
print(random.random())
```

```python
random.randrange(10) # escolhe aleatoriamente de range(10)
```

```python
random.randrange(3, 6) # escolhe aleatoriamente de range(3, 6)
```

```python
# reordena os elementos de uma lista aleatoriamente
up_to_ten = list(range(10))
random.shuffle(up_to_ten)
print(up_to_ten)
```

```python
# escolhe um elemento randomicamente de uma lista
my_best_friend = random.choice(['Alice', 'Bob', 'Charlie'])

my_best_friend
```

```python
# escolhe aleatoriamente uma amostra dos elementos sem substituição (por exemplo, sem duplicatas)
lottery_numbers = range(25)
winning_numbers = random.sample(list(lottery_numbers), 15)

winning_numbers
```

```python
# escolhe uma amostra de elementos com substituição (por exemplo, permitindo duplicatas)
[random.choice(range(10)) for _ in range(4)]
```
