# Ordenação

```python
x = [4, 1, 2, 3]
y = sorted(x) # é [1, 2, 3, 4], x não mudou
x.sort() # agora x é [1, 2, 3, 4]
```

```python
# Organiza a lista pelo valor absoluto do maior para o menor
x = sorted([-4, 1, -2,3], key=abs, reverse=True) # is [-4, 3, -2,1]
x
```
