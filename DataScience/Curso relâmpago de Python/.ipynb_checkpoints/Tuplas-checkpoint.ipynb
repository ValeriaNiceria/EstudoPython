# Tuplas

```python
my_list = [1, 2]
my_list
```

```python
my_tuple = (1, 2)
my_tuple
```

```python
other_tuple = 1, 2
other_tuple
```

```python
my_list[1] = 3
my_list
```

```python
try:
    my_tuple[1] = 3
except TypeError:
    print("Cannot modify a tuple")
```

```python
def sum_and_product(x, y):
    return (x + y), (x * y)

sum_and_product(2, 3)
```

```python
s, p = sum_and_product(5, 10)
```
